```python
df['rating']=1
df_rating=df.groupby('post_id',as_index=False).sum()
```

```python
df_rating.dtypes
```

```python
post_ids2=df_rating['post_id'].tolist()
rating2=df_rating['rating'].tolist()
rating_mapping=dict(zip(post_ids2,rating2))  #makes a dictionary with post_id2 as keys and rating values as value

df['rating']=df['post_id'].map(rating_mapping)
display(df)
```

```python
merged_df = pd.merge(df, df1, on='post_id', how='inner')
merged_df
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the label encoder on your column
merged_df['user_encoded'] = label_encoder.fit_transform(merged_df['user_id'])
merged_df['post_encoded'] = label_encoder.fit_transform(merged_df['post_id'])

merged_df.head()
```

```python
users_pivot= merged_df.pivot_table(index=['title'], columns=['user_encoded'], values=['rating'])
users_pivot.fillna(0, inplace=True)
users_pivot
```

```python
from scipy.sparse import csr_matrix

users_pivot_matrix = csr_matrix(users_pivot.values)

from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(users_pivot_matrix)
```

```python
users_pivot.shape
```

```python
import numpy as np
query_index = np.random.choice(users_pivot.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(users_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(users_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, users_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```
